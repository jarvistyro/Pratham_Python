```python
import speech_recognition as sr
import os
import pyttsx3

r = sr.Recognizer()
def pyhelp():
    print("Let me explain...")
    engine = pyttsx3.init()
    engine.say("Let me explain...")
    engine.runAndWait()
    print("I am here to full fill your wishes you just have to say what you want..")
    engine = pyttsx3.init()
    engine.say("I am here to full fill your wishes you just have to say what you want..")
    engine.runAndWait()
    #print("like.....You can type... play music.... I will take to place where you can choose and play the music of your choice")
    print("Say... play music....")
    engine = pyttsx3.init()
    engine.say("like.....You can say... play music.... your Ginnie will take to place where you can choose and play the music of your choice")
    engine.runAndWait()
    print("Or")
    engine = pyttsx3.init()
    engine.say("or")
    engine.runAndWait()
    print("say... I want to draw.... ")
    engine = pyttsx3.init()
    engine.say("You can say... I want to draw.... and your Ginnie will take you to a place where you can draw and paint")
    engine.runAndWait()
    engine = pyttsx3.init()
    engine.say("And when your wishes get fulfilled or you get bored    you can simply say")
    engine.runAndWait()
    print(" I want to go out.... ")
    engine = pyttsx3.init()
    engine.say("I want to go out.... and your Ginnie will send you back to your world")
    engine.runAndWait()
    print("Ready.... ")
    engine = pyttsx3.init()
    engine.say("We are good to go now.. i am eagerly waiting for see your wishes my Lord")
    engine.runAndWait()
    wish()

def check(j):
    if ("yes" in j) or ("tell me" in j) or ("explain" in j):
        print("   ")
        pyhelp()
    elif ("no" in j) or ("i know" in j) or ("dont explain" in j):
        print("Great..")
        engine = pyttsx3.init()
        engine.say("Great   Feels good to meet smart people like you..")
        engine.runAndWait()
        wish()
    elif ("exit" in j) or ("quit" in j) or ("leave" in j) or ("bye" in j) or ("out" in j):
        print("Hope to see you soon.. Bye Bye")
        engine = pyttsx3.init()
        engine.say("Hope to see you soon.. Bye Bye")
        engine.runAndWait()
        #a=False
        
    else:
        print("ok")
    
    

        
    
    
def wish():
    while True:
        print("Make a wish: ")
        engine = pyttsx3.init()
        engine.say("make a wish")
        engine.runAndWait()
        with sr.Microphone() as source:
            audio = r.listen(source)
            p = r.recognize_google(audio)
            if (("don't" in p) or ("dont" in p) or ("do not" in p)) and ("run" in p):
                print("Oops..")
                engine = pyttsx3.init()
                engine.say("Oops  you have just asked me to dont run.. ")
                engine.runAndWait()
            elif ("explain" in p) or ("forgot" in p) or ("how to" in p) or ("what should i do" in p):
                engine = pyttsx3.init()
                engine.say("Ooh Sure")
                engine.runAndWait()
                pyhelp()
            elif ("world" in p) or ("chrome" in p) or ("internet" in p):
                engine = pyttsx3.init()
                engine.say("Why not")
                engine.runAndWait()
                os.system("chrome")
                
            elif ("run" in p) or ("notepad" in p) or ("editor" in p) or ("write" in p):
                engine = pyttsx3.init()
                engine.say("Here you go")
                engine.runAndWait()
                os.system("notepad")
                
            elif ("draw" in p) or ("paint" in p) or ("colour" in p) or ("sketch" in p):
                engine = pyttsx3.init()
                engine.say("Certainly my lord")
                engine.runAndWait()
                os.system("mspaint")
                
            elif ("play" in p) or ("music" in p) or ("songs" in p) or ("sound" in p):
                engine = pyttsx3.init()
                engine.say("of course")
                engine.runAndWait()
                os.system("wmplayer")
                
            elif ("exit" in p) or ("quit" in p) or ("leave" in p) or ("bye" in p) or ("out" in p):
                print("Hope to see you soon.. Bye Bye")
                engine = pyttsx3.init()
                engine.say("Hope to see you soon.. Bye Bye.. Do come again")
                engine.runAndWait()
                break    
            else:
                print("My Appologies.. ")
                engine = pyttsx3.init()
                engine.say("My Appologies.. Aaka I am not allowed to do this")
                engine.runAndWait()
                print("You can try somthing else")
                engine = pyttsx3.init()
                engine.say("but You can try somthing else")
                engine.runAndWait()
        
            
print("*******************************************************************************")
print("             Welcome to the world of Ginnie.... Koi Hukum mere AAKA")
print("*******************************************************************************")
engine = pyttsx3.init()
engine.say("Welcome !!!!    to the world of Ginnie... KOI HUKUM MERE AAAKA")
engine.runAndWait()
print("=> You know the rules?? or want me to explain??")
engine = pyttsx3.init()
engine.say("You know the rules or want me to explain   my lord")
engine.runAndWait()

with sr.Microphone() as source: 
    audio = r.listen(source)
    j = r.recognize_google(audio)
    check(j)
```
